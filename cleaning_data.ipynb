```python
import pandas as pd
```

```python
data = pd.read_csv('data_before.csv')
data.head()
```

```python
#dropping the last 2 columns:
data.drop(data.iloc[:,-2:], axis=1, inplace=True)
data.head()
```

```python
#dropping complete rows with NA value:
data = data[data['Row Type'].notna()]
data.head(15)
```

```python
#Adding an identity column (iter for iteration):
lst = []
counter = 0
for i in data['Row Type']:
    if 'first name' in i:
        counter+=1 
    lst.append(counter)
```

```python
data['iter'] = lst
data.head()
```

## To 2 DataFrames:

```python
#splitting the data frame into two:
frst = data[data['Row Type'].str.contains('first name')]
scnd = data[~data['Row Type'].str.contains('first name')]
```

```python
scnd = scnd[~scnd['Row Type'].str.contains('Row Type')]
scnd[:20]
```

## Wrangling 1st dataframe:

```python
frst.head()
```

```python
#selecting unused columns:
frst.iloc[:,3:-1]
```

```python
#dropping them:
frst = frst.drop(frst.iloc[:,3:-1], axis=1)
```

```python
#renaming cols:
frst.rename(columns={'Row Type':'first_name',
                        'Iter Number':'last_name',
                        'Power1':'date'}, inplace=True)
```

```python
#extracting extra text in fields:
clean_func = lambda x: x[x.index(': ')+1:]
frst['first_name'] = frst['first_name'].map(clean_func)
frst['last_name'] = frst['last_name'].map(clean_func)
frst['date'] = frst['date'].map(clean_func)
frst
```

## Merging into one dataframe:

```python
final_result = pd.merge(frst, scnd, on='iter')
final_result
```

```python
final_result.to_csv('data_after.csv', header=True, index=False)
```
